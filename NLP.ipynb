```python
text = "I am feeling very happy today!!!..."
print("Original text:",text)
```

```python
text = text.lower() #Step 1: Convert to lowercase because "Happy" and "happy" should be treated the same so ML will consider these two as different words.
print("Lowercase text:",text)
```

```python
#step 2: Remove punctuation as we remove everything except letters and spaces
import re #regular expression library 
text= re.sub(r'[^\w\s]','',text) #\w means word character, \s means space, ^ means not, so this regex means remove everything except word characters and spaces
print("Text without punctuation:",text)
```

```python
#Step 3: Tokenization: Split the text into individual words (tokens)
tokens= text.split()
print("Tokens: ",tokens)
```

```python
#step 4:remove stopwords: Stopwords are common words that do not carry much meaning and can be removed to reduce noise in the data. Examples of stopwords include "the", "is", "in", "and", etc.
stopwords = ["i","am","is","in","and","the","a","an","to","of","very"] 
filtered_tokens=[]
for word in tokens:
    if(word not in stopwords):
        filtered_tokens.append(word)

print("Filtered tokens: ",filtered_tokens)
```

```python
def stem(word):
    if word.endswith("ing"):
        return word[:-3] #remove "ing" from the end of the word
    return word

stemmed_words=[]
for word in filtered_tokens:
    stemmed_words.append(stem(word))

print("Stemmed words: ",stemmed_words)
```

```python
from sklearn.feature_extraction.text import CountVectorizer #will help us to convert text to count
```

```python
sentences = [
    "I am good","i am happy", "happy today"," i am sad"
]
```

```python
#vectorize
vectorizer = CountVectorizer() #create an object of CountVectorizer
x= vectorizer.fit_transform(sentences) #learn vocabulary and convert sentences to numbers
```

```python
print("Vocabulary:",vectorizer.get_feature_names_out()) #returns the vocabulary learned from the sentences
```

```python
print("Vectors:")
print(x.toarray()) 
#Here each row = sentence and each column = word from vocabulary
#Each value= count of the word in the sentence
```

```python
import nltk
#4 makor uses of nltk
#1. Tokenization    
#2. Stopword removal
#3. Stemming   
#4 Lemmatization

nltk.download('punkt_tab') #download punkt tokenizer
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
from nltk.tokenize import word_tokenize #for tokenization
from nltk.corpus import stopwords #for stopword removal
from nltk.stem import PorterStemmer, WordNetLemmatizer #for stemming and lemmatization
```

```python
text= "I am learning python programming, it is very helpful!!"
print("Original text:",text)
```

```python
#step 1:lowercase
text= text.lower()
print("Lowercase text:",text)
```

```python
#step 2: tokenization
tokens= word_tokenize(text)
print("Tokens:",tokens)
```

```python
#step 3: punctuation removal
import string
punc= string.punctuation
print("Punctuation:",punc)
```

```python
punctuation_filter= [word for word in tokens if word not in punc]
print("Tokens without punctuation:",punctuation_filter)
```

```python
#step 4: stopword removal
eng_stopwords= stopwords.words('english')
print("English Stopwords:",eng_stopwords)
filtered_tokens= [word for word in punctuation_filter if word not in eng_stopwords]
print("Tokens after stopword removal:",filtered_tokens)
```

```python
#step 5: Stemming & Lemmatization
stem=PorterStemmer()
stem.stem("went")
```

```python
wnet = WordNetLemmatizer()
print(wnet.lemmatize("playing","v")) #lemmatize the word "playing" as a verb, it will return "play" because the base form of "playing" is "play"
print(wnet.lemmatize("flying","v")) #lemmatize the word "flying" as a verb, it will return "fly" because the base form of "flying" is "fly"
print(wnet.lemmatize("went","v")) #lemmatize the word "went" as a verb, it will return "go" because the base form of "went" is "go"
print(wnet.lemmatize("bought","v")) #lemmatize the word "bought" as a verb, it will return "buy" because the base form of "bought" is "buy")

```

```python
lemmatized_words= []
for word in filtered_tokens:
    lemmatized_words.append(wnet.lemmatize(word,"v"))
print("Lemmatized words:",lemmatized_words)
```

```python
cleaned_text= " ".join(lemmatized_words)
print("Cleaned text:",cleaned_text)
```

```python
sentences= ["I am good","i am happy", "happy today"," i am sad"]
def preprocess_text(sentences):
    cleaned_sentences= []
    for sentence in sentences:
        
            sentence= sentence.lower()
            #step 2: tokenization
            tokens= word_tokenize(sentence)
            #step 3: punctuation removal
            punctuation_filter= [word for word in tokens if word not in punc]
            #step 4: stopword removal
            filtered_tokens= [word for word in punctuation_filter if word not in stopwords.words('english')]
            #step 5: Stemming & Lemmatization
            lemmatized_words= []
            for word in filtered_tokens:
                lemmatized_words.append(wnet.lemmatize(word,"v"))
            cleaned_sentence= " ".join(lemmatized_words)
            cleaned_sentences.append(cleaned_sentence)
    return cleaned_sentences

cleaned_text= preprocess_text(sentences)
print("Cleaned sentences:",cleaned_text)
```
